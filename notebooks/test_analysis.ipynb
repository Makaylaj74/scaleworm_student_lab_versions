```python
import zipfile
import os

zip_path = "/home/jovyan/scaleworm-student-lab/notebooks/verification_session/verified_scaleworm_dataset_2024-10-01_to_2024-10-31.zip"
extract_path = "/home/jovyan/scaleworm_student_lab/notebooks/verification_session/"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Unzipped!")
```

```python
import os

base_path = extract_path

for root, dirs, files in os.walk(base_path):
    print(root)
    break
```

```python
import glob
import pandas as pd

label_path = os.path.join(base_path, "labels")  # adjust if needed

rows = []

for file in glob.glob(label_path + "/*.txt"):
    frame_id = os.path.basename(file).replace(".txt", "")
    
    with open(file, "r") as f:
        lines = f.readlines()
    
    for line in lines:
        parts = list(map(float, line.strip().split()))
        
        rows.append({
            "frame_id": frame_id,
            "class": int(parts[0]),
            "x_center": parts[1],
            "y_center": parts[2],
            "width": parts[3],
            "height": parts[4]
        })

df = pd.DataFrame(rows)

print(df.head())
print("Total detections:", len(df))
```

```python
from datetime import datetime

def extract_timestamp(frame_id):
    try:
        # adjust this depending on your filename format
        time_str = frame_id.split("-")[1]
        return datetime.strptime(time_str, "%Y%m%dT%H%M%SZ")
    except:
        return None

df["timestamp"] = df["frame_id"].apply(extract_timestamp)

df = df.dropna(subset=["timestamp"])
```

```python
worms_per_time = df.groupby("timestamp").size()

import matplotlib.pyplot as plt

plt.figure()
worms_per_time.plot()
plt.title("Scale Worm Observations Over Time")
plt.xlabel("Time")
plt.ylabel("Count")
plt.show()
```

```python
df["time_bin"] = df["timestamp"].dt.floor("H")

worms_hourly = df.groupby("time_bin").size()

plt.figure()
worms_hourly.plot()
plt.title("Hourly Scale Worm Abundance")
plt.xlabel("Time")
plt.ylabel("Count")
plt.show()
```
